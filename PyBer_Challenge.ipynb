# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_city_type = pyber_data_df.groupby('type')['ride_id'].count()
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_city_type = city_data_df.groupby('type')['driver_count'].sum()
total_drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_city_type = pyber_data_df.groupby('type')['fare'].sum()
total_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_by_city_type = pyber_data_df.groupby('type')['fare'].mean()
average_fare_by_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver_by_city_type = total_fares_by_city_type / total_drivers_by_city_type
average_fare_per_driver_by_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_series_list = [total_rides_by_city_type, total_drivers_by_city_type, total_fares_by_city_type, average_fare_by_city_type, average_fare_per_driver_by_city_type]
pyber_summary_df = pd.concat(pyber_series_list, axis=1)
pyber_summary_df.columns = ['Total Rides', 'Total Drivers', 'Total Fares', 'Average Fare per Ride', 'Average Fare per Driver']
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].apply(lambda x: f"${x:.2f}")
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].apply(lambda x: f"${x:.2f}")
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].apply(lambda x: f"${x:.2f}")
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
weekly_fares_df = pyber_data_df.groupby(['type', 'date'])[['fare']].sum()
weekly_fares_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
weekly_fares_df.reset_index(inplace=True)
weekly_fares_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
weekly_pivot = weekly_fares_df.pivot(index='date', columns='type', values='fare')
weekly_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
weekly_pivot_filtered = weekly_pivot['2019-01-01':'2019-04-29']
weekly_pivot_filtered
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

weekly_pivot_filtered.index = pd.to_datetime(weekly_pivot_filtered.index)
weekly_pivot_filtered
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
weekly_pivot_filtered.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_pivot_resampled = weekly_pivot_filtered.resample('W').sum()
weekly_pivot_resampled
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
# Plot graph
fig = weekly_pivot_resampled.plot(figsize=(20,5))
fig.set_title('Weekly Fares by City Type')
fig.set_xlabel('Date')
fig.set_ylabel('Sum of Fares')
fig.legend(loc=1)
plt.savefig('Weekly Fares by City Type.png', bbox_inches='tight')
```
