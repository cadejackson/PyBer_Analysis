```python
# Concat dfs to one df
df = pd.concat([df1,df2,df3,df4])
```

```python
# Get row count
len(df)
```

```python
# Check min and max for each hole size
df.groupby('Hole Diameter (in)').agg({'Hole Depth (feet)': [np.min, np.max]}).sort_index(ascending=False)
```

```python
# Correct 17.5" Hole Size
df.loc[(df['Hole Depth (feet)'] >= 0) & (df['Hole Depth (feet)'] <= 382), 'Hole Diameter (in)'] = 17.5
```

```python
# Correct 12.25" Hole Size
df.loc[(df['Hole Depth (feet)'] > 382) & (df['Hole Depth (feet)'] <= 3494), 'Hole Diameter (in)'] = 12.25
```

```python
# Correct 8.75" Hole Size
df.loc[(df['Hole Depth (feet)'] > 3494) & (df['Hole Depth (feet)'] <= 9145), 'Hole Diameter (in)'] = 8.75
```

```python
# Check min and max for each hole size
df.groupby('Hole Diameter (in)').agg({'Hole Depth (feet)': [np.min, np.max]}).sort_index(ascending=False)
```

```python
# Review -999.25 values
df[df['Hole Diameter (in)'] == -999.25]
```

```python
# Drop rows with -999.25 for bit diameter
df = df[df['Hole Diameter (in)'] != -999.25]
```
